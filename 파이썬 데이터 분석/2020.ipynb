```python
#df['사고일시'] = pd.to_datetime(df['사고일시'])
df['year'] = df.사고일시.str.split(' ').str[0]
df['month'] = df.사고일시.str.split(' ').str[1]
df['day'] = df.사고일시.str.split(' ').str[2]
df['time'] = df.사고일시.str.split(' ').str[3]
```

```python
df
```

```python
df_화물 = df[(df['가해운전자 차종']=='화물') | (df['피해운전자 차종']=='화물')|(df['사고내용']=='사망사고')]
display(df_화물['time'].value_counts())
```

```python
print(df_화물.groupby("time")['사망자수'].sum().sort_values(ascending = False).head(10))
```

```python
df_화물사망 = pd.DataFrame(df_화물.groupby("time")['사망자수'].sum(), columns=['사망자수'])
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(20, 5))
plt.plot(df_화물사망['사망자수'],color='green',marker='o',linestyle='solid')
plt.title('시간대별 사망자수')
plt.ylabel("사망자수")
plt.show()
```
